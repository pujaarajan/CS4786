```python
import numpy
import math
import scipy.linalg
```

```python
import matplotlib.pyplot
%matplotlib inline
```

```python
a1 = numpy.random.normal(0, 1, (1000,1))
b1 = numpy.random.normal(0, math.sqrt(2), (1000,1))
X = numpy.concatenate((a1,b1), axis = 1)
```

```python
matplotlib.pyplot.scatter(a1,b1)
```

```python
rotation_matrix = [[1/math.sqrt(2), -1/math.sqrt(2)],[1/math.sqrt(2),1/math.sqrt(2)]]
```

```python
rotation_matrix
```

```python
rotated_matrix = numpy.dot(X, rotation_matrix)
```

```python
X2 = numpy.dot(X, rotation_matrix)+1
```

```python
matplotlib.pyplot.scatter(a1,b1,color='red')
matplotlib.pyplot.scatter(X2[:,0],X2[:,1],color='blue')
```

#PCA of Pre Rotation

```python
covariance_matrix = numpy.cov(X[:,0],X[:,1])
```

```python
W = numpy.linalg.eigh(covariance_matrix)[1]
```

```python
Y = numpy.dot(X, W)
```

```python
matplotlib.pyplot.scatter(Y[:,0],Y[:,1])
```

#PCA of Post Rotation

```python
covariance_matrix2 = numpy.cov(X2[:,0],X2[:,1])
```

```python
W2 = numpy.linalg.eigh(covariance_matrix2)[1]
```

```python
Y2 = numpy.dot(X2, W2)
```

```python
matplotlib.pyplot.scatter(Y2[:,0],Y2[:,1])
```
